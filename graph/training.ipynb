- 从第三个数开始，每个数是前两个数的和。

**任务要求**：

2. 使用 `for` 循环调用生成器函数，打印出前 `n` 个斐波那契数。

---

```python
    print(num)
```

输出：

#### 示例 2：
调用：

```python
for num in fibonacci(10):
    print(num)
```

输出：
```plaintext

1
2
3
5
8
13
21
34
```

---

### 提示：

- 使用两个变量 `a` 和 `b` 表示前两个斐波那契数，初始化为 `0` 和 `1`。
- 每次生成一个斐波那契数后，更新 `a` 和 `b` 的值：`a, b = b, a + b`。

````python
def fibonacci(n):
    list = []
    for i in range(n):
        if i <= 1: 
            list.append(i)
            yield i
        
        else:
            s = list[i-1]+list[i-2]
            list.append(s)
            yield s


````

### 使用 `yield from` 合并多个生成器

编写一个生成器函数 `chain_generators(gen1, gen2)`，它将两个生成器 `gen1` 和 `gen2` 的输出合并到一起，使用 `yield from` 来实现合并。

**要求**：
1. 实现生成器函数 `chain_generators(gen1, gen2)`。
2. 创建两个生成器：
   - `gen1`：生成 1 到 3 的数字。
   - `gen2`：生成 'a' 到 'c' 的字符。
3. 调用 `chain_generators(gen1, gen2)`，依次输出两个生成器的内容。

---

### 输出示例：
调用以下代码：
```python
def gen1():
    for i in range(1, 4):
        yield i

def gen2():
    for char in 'abc':
        yield char

for item in chain_generators(gen1(), gen2()):
    print(item)
```

**输出**：
```plaintext
1
2
3
a
b
c
```

---

### 提示：
- `yield from` 可以直接委托生成器的生成任务，而无需手动迭代。
- 可以将 `yield from` 分别用在 `gen1` 和 `gen2` 的生成逻辑中。

````python
def chain_generators(gen1,gen2):
    yield from gen1()
    yield from gen2()

def gen1():
    for i in range(3):
        yield i+1

def gen2():
    for char in 'abc':
        yield char

        
for item in chain_generators(gen1, gen2):
    print(item)
````

完成后，你将能够使用生成器优雅地实现图的深度优先遍历！

````python
def dfs(graph, start, visited=None):
    """
    使用生成器实现深度优先遍历
    :param graph: 图的邻接表表示
    :param start: 起始节点
    :param visited: 已访问节点的集合
    """
    # 实现 DFS
    visited = set()
    def _helper(graph,node):
        if node in visited: return
        visited.add(node)
        yield node
        
        for n in graph[node]:
            yield from _helper(graph,n)
    
    yield from _helper(graph, start)

# 示例图
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# 调用生成器并打印节点
for node in dfs(graph, 'A'):
    print(node)
````
